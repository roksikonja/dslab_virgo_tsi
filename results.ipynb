```python
import os
import pickle
import json

import pandas as pd
import numpy as np

from dslab_virgo_tsi.base import FitResult, FinalResult, OutResult, BaseSignals
from dslab_virgo_tsi.constants import Constants as Const
from dslab_virgo_tsi.visualizer import Visualizer
```

```python
visualizer = Visualizer()
visualizer.set_figsize()
```

```python
# Load data
results_dir = "./results/results/2019-12-17_01-47-09_smooth_monotonic_svgp_FINAL_m1000"
with open(os.path.join(results_dir, "SMOOTH_MONOTONIC_modeling_result.pkl"), "rb") as f:
    result_ = pickle.load(f)
```

```python
print(result_.__dict__.keys())
```

```python
# BaseSignals - Input data
base_sig: BaseSignals = result_.base_signals

# print(base_sig.__dict__)
print(base_sig.__dict__.keys())

visualizer.plot_signals(
    [
        (base_sig.t_a_nn, base_sig.a_nn, f"{Const.A}_raw", False),
        (base_sig.t_b_nn, base_sig.b_nn, f"{Const.B}_raw", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, y_lim=[1357, 1369], legend="upper right", 
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, base_sig.a_mutual_nn, f"{Const.A}_raw_mutual", False),
        (base_sig.t_mutual_nn, base_sig.b_mutual_nn, f"{Const.B}_raw_mutual", False),
    ],
results_dir=None, title=None,x_ticker=Const.XTICKER, y_lim=[1357, 1369], legend="upper right", 
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)

visualizer.plot_signals(
    [
        (base_sig.t_a_nn, base_sig.exposure_a_nn, f"{Const.A}_raw_exposure", False),
        (base_sig.t_b_nn, base_sig.exposure_b_nn, f"{Const.B}_raw_exposure", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right", 
x_label=Const.YEAR_UNIT, y_label="EXPOSURE")

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, base_sig.exposure_a_mutual_nn, f"{Const.A}_raw_exposure_mutual", False),
        (base_sig.t_mutual_nn, base_sig.exposure_b_mutual_nn, f"{Const.B}_raw_exposure_mutual", False),
    ],
results_dir=None, title=None,x_ticker=Const.XTICKER, legend="upper right", 
x_label=Const.YEAR_UNIT, y_label="EXPOSURE")
```

```python
# FitResult
before_fit: FitResult = result_.history_mutual_nn[0]

print(before_fit.__dict__.keys())

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, before_fit.a_mutual_nn_corrected, f"{Const.A}_mutual_nn", False),
        (base_sig.t_mutual_nn, before_fit.b_mutual_nn_corrected, f"{Const.B}_mutual_nn", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, before_fit.ratio_a_b_mutual_nn_corrected, f"ratio_mutual_nn", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)
```

```python
# FitResult
last_iter: FitResult = result_.history_mutual_nn[-1]
    
print(last_iter.__dict__.keys())

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, last_iter.a_mutual_nn_corrected, f"{Const.A}_mutual_nn", False),
        (base_sig.t_mutual_nn, last_iter.b_mutual_nn_corrected, f"{Const.B}_mutual_nn", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)

visualizer.plot_signals(
    [
        (base_sig.t_mutual_nn, last_iter.ratio_a_b_mutual_nn_corrected, f"ratio_mutual_nn_last", False),
        (base_sig.t_mutual_nn, before_fit.ratio_a_b_mutual_nn_corrected, f"ratio_mutual_nn_first", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT)
```

```python
# FinalResult
final_res: FinalResult = result_.final

print(final_res.__dict__.keys())

visualizer.plot_signals(
    [
        (base_sig.t_a_nn, final_res.degradation_a_nn, f"DEGRADATION_{Const.A}", False),
        (base_sig.t_b_nn, final_res.degradation_b_nn, f"DEGRADATION_{Const.B}", False)
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.DEGRADATION_UNIT)

visualizer.plot_signals(
    [
        (base_sig.t_a_nn, final_res.a_nn_corrected, f"{Const.A}_corrected", False),
        (base_sig.t_b_nn, final_res.b_nn_corrected, f"{Const.B}_corrected", False),
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper right",
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT, y_lim=[1362, 1369])
```

```python
# OutResult
out_res: OutResult = result_.out

print(out_res.__dict__.keys())
print(out_res.params_out.__dict__.keys())

visualizer.plot_iter_loglikelihood(out_res.params_out.svgp_iter_loglikelihood, results_dir=None,
title=None, legend="lower left", x_label=Const.ITERATION_UNIT, y_label=Const.LOG_LIKELIHOOD_UNIT)

visualizer.plot_signals_mean_std_precompute(
    [
        (out_res.t_hourly_out, out_res.signal_hourly_out, out_res.signal_std_hourly_out, f"TSI_hourly")
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper left", y_lim=[1362, 1369],
x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT, max_points=1e6,
inducing_points=out_res.params_out.svgp_inducing_points)

# Prior
visualizer.plot_signals_mean_std_precompute(
    [
        (out_res.t_hourly_out, out_res.signal_hourly_out, out_res.signal_std_hourly_out,
         f"TSI_hourly")
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper left",
y_lim=[1362, 1369],x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT, max_points=1e7,
f_sample_triplets=[(out_res.params_out.svgp_t_prior, out_res.params_out.svgp_prior_samples, "PRIOR")])


# Posterior
visualizer.plot_signals_mean_std_precompute(
    [
        (out_res.t_hourly_out, out_res.signal_hourly_out, out_res.signal_std_hourly_out,
         f"TSI_hourly")
    ],
results_dir=None, title=None, x_ticker=Const.XTICKER, legend="upper left",
y_lim=[1362, 1369], x_label=Const.YEAR_UNIT, y_label=Const.TSI_UNIT, max_points=1e7,
f_sample_triplets=[(out_res.params_out.svgp_t_posterior, out_res.params_out.svgp_posterior_samples, "POSTERIOR")])
```
